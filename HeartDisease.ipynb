importing required libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing

```python
heart_data = pd.read_csv('heart.csv')
```

```python
#print first 5 rows
heart_data.head()
```

```python
# print last 5 rows 
heart_data.tail()
```

```python
heart_data.describe()
```

```python
# cheking the value of the target
heart_data['target'].value_counts()
```

0 - Healthy heart
1 - Diseased heart

splitting the features and target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training using Logistic regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
testing_accuracy
```

building a predictive system

```python
input_data = (52,1,3,118,186,0,0,190,0,0,1,0,1)
```

```python
data_as_numpy_array = np.asarray(input_data)

shaped_array = data_as_numpy_array.reshape(1,-1)
```

```python
prediction = model.predict(shaped_array)

if(prediction == 1):
  print("the patient has heart disease")
else:
  print("the patient id healthy")
```
